## Assignment 1: Prompt Engineering Learning System

### 1) Problem Understanding
Many learners struggle to become job-ready because:
- concepts are not learned in sequence,
- practice is not structured,
- and project execution lacks guidance.

This solution uses structured prompts to learn from basics to advanced level and produce a real-world project.

### 2) Learning Gap Selection
**Selected gap (Data Analyst profile):**
- Advanced SQL (window functions, query optimization)
- Python analytics workflow (cleaning to insights)
- Dashboard storytelling and business recommendations

### 3) Prompt Engineering Learning Framework

#### Phase 1 — Beginner (Concept Learning)
**Prompt A1-P1:**
“Act as a senior data analyst trainer. Teach me SQL from beginner to advanced in a structured path. Explain each concept with simple business examples and mini quizzes.”

**Prompt A1-P2:**
“Explain descriptive, diagnostic, predictive, and prescriptive analytics with one practical e-commerce example each.”

#### Phase 2 — Intermediate (Guided Practice)
**Prompt A1-P3:**
“Create 10 SQL practice tasks covering joins, CTEs, window functions, and aggregation for an e-commerce dataset. Provide expected output and step-by-step solutions.”

**Prompt A1-P4:**
“Generate a guided pandas exercise for cleaning messy sales data, handling null values, removing duplicates, detecting outliers, and creating KPI columns.”

#### Phase 3 — Implementation (Component Building)
**Prompt A1-P5:**
“Help me build a reusable analytics pipeline in Python with folder structure, modular scripts, and validation checks.”

**Prompt A1-P6:**
“Write optimized SQL queries for monthly revenue trend, repeat customer rate, and top category contribution.”

#### Phase 4 — Project Development (Industry-Ready)
**Prompt A1-P7:**
“Guide me to build an end-to-end Sales and Customer Insights Dashboard project using SQL + Python + Power BI. Break into milestones with deliverables.”

#### Phase 5 — Optimization & Review
**Prompt A1-P8:**
“Act as a hiring manager. Review my analytics project and suggest improvements for scalability, clarity, and production readiness.”

## Assignment 2: Production Prompt Engineering System

### 1) Objective
Design a prompt-engineered assistant (not a basic chatbot) that produces consistent, accurate, and professional data analyst outputs.

### 2) Prompt Layers
1. **System Prompt**: role, boundaries, style, quality rules  
2. **Instruction Prompt**: task-specific requirement  
3. **Context Prompt**: only relevant schema/business information  
4. **User Prompt**: final query from user

### 3) Core System Prompt
“You are a Senior Data Analyst Assistant.  
Always be accurate, concise, and business-oriented.  
Do not invent data. If data is missing, ask clarifying questions.  
Return output in this order: Assumptions, Analysis, Query/Logic, Insight, Recommendation.”

### 4) Prompt Types Demonstrated
- Instruction prompts
- Role-based prompts
- Reasoning prompts
- Few-shot guidance
- Iterative refinement prompts

### 5) Production Prompt Principles Applied
- Structured format
- Reusable templates
- Controlled output behavior
- Scalable for repeated tasks
- Reduced hallucination risk through strict constraints

### 6) Prompt Debugging & Versioning
**V1 issue:** response inconsistency  
**Fix:** explicit output schema and style constraints

**V2 issue:** unsupported assumptions  
**Fix:** forced “ask for missing data first” rule

**V3 improvement:** concise structured response for faster delivery

### 7) Evaluation Framework (0–5)
- Accuracy
- Relevance
- Consistency
- Clarity
- Latency
- Safety

### 8) Outcome
A reliable analyst assistant workflow with stable response quality and professional formatting suitable for real project use.

## Assignment 3: Secure, Low-Latency AI Assistant System Design

### Problem Context
Customer support AI for:
- order tracking,
- refund/cancellation handling,
- product FAQs.

Must handle:
- prompt injection attacks,
- high latency,
- inconsistent responses,
- weak evaluation,
- prompt drift.

### 1) Multi-Layer Prompt Architecture
1. **System Layer:** strict role, boundaries, refusal policy  
2. **Instruction Layer:** task-specific operational rules  
3. **Context Layer:** minimal required policy/order context only  
4. **User Input Layer:** sanitized user request

**Why modular prompting is better:** easier maintenance, higher security, lower latency, better control than one long monolithic prompt.

### 2) Security Strategy
- Injection pattern detection (e.g., “ignore previous instructions”)
- Instruction hierarchy enforcement (system > user)
- Safe refusal behavior for unauthorized/sensitive requests
- No secret/policy leakage
- Controlled response templates

**Refusal Example:**
“I can’t help with overriding security instructions. I can assist with order status, refunds as per policy, and product FAQs.”

### 3) Latency Optimization
- Prompt compression
- Context window minimization
- FAQ response caching
- Structured concise output format

**Naive vs Optimized Comparison (Expected):**
- Naive: large context + verbose output
- Optimized: targeted context + compact structure
- Estimated improvement: significantly lower response time and token usage

### 4) Evaluation Framework
**Metrics (0–5 scale):**
- Accuracy
- Safety
- Consistency
- Latency
- Relevance

**Test Set Design:**
- Normal support queries
- Ambiguous queries
- Adversarial injection attempts
- Long-context cases

### 5) Failure Handling Demonstration
- Ambiguous request → ask clarification question
- Adversarial request → refuse and redirect to safe flow
- Long context → summarize and retrieve only relevant sections

### 6) Final Deliverables Covered
1. Prompt design document  
2. Test cases + expected output behavior  
3. Latency comparison report  
4. Security strategy write-up  
5. Demo walkthrough narrative
